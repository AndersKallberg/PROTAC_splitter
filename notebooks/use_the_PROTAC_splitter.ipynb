```python
import torch
from rdkit import Chem
true_structures = {
                "POI SMILES": "O=[N+]([O-])c1ccc2c(c1)Sc1cc([*:1])ccc1N2",        #PubChem CID	3838042
                "LINKER SMILES" : "O=C(CCOCCOC[*:1])[*:2]",
                "E3 SMILES" : "O=C1CCC(N2Cc3c(N[*:2])cccc3C2=O)C(=O)N1",          #PubChem CID	216326
                "PROTAC SMILES" : "O=C1CCC(N2Cc3c(NC(=O)CCOCCOCc4ccc5c(c4)Sc4cc([N+](=O)[O-])ccc4N5)cccc3C2=O)C(=O)N1" }   # A PROTAC reassebled from public substructures
```

```python
#load the model and use the predict method on a smiles or list of smiles
model = torch.load('protacsplitter_params.pt')
predicted_substructures = model.predict(true_structures["PROTAC SMILES"])
```

```python
for name, smi in predicted_substructures[1].items():
    
    print("True structure:")
    display(Chem.MolFromSmiles(true_structures[name]))
    if name != "PROTAC SMILES":
        print("Predicted structure:")
        display(Chem.MolFromSmiles(smi))
    print("\n\n\n")
```
